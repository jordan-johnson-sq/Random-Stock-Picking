```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats as st
```

```python
import statistics
```

```python
rng = np.random.default_rng()
```

```python
rd = rng.normal(loc=10, scale=7, size=10000)
```

```python
dd = rng.normal(loc=10, scale=2, size=10000)
```

```python
plt.hist(rd, bins=100, alpha=0.3)
plt.hist(dd, bins=100, alpha=0.5)
plt.x
plt.show()
```

```python
s = np.std(rd)
```

```python
m = np.mean(rd)
```

```python
z = statistics.NormalDist(m, s).zscore(5) 
```

```python
st.norm.ppf(z)
```

```python
bootstrapped_means = np.array([])

for i in range(2000):
    bootstrapped_sample = np.random.choice(rd, size=5, replace=True)
    mean = bootstrapped_sample.mean()
    bootstrapped_means.append(mean)
```

```python
plt.hist(bootstrapped_means, bins=100)
plt.show()
```

```python
bootstrapped_means.std()
```
